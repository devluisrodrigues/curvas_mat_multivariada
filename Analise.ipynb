````python
def encontraRaiz(valores):
    todos = []
    # percorre todos os t em valores
    # analisa se o valor de f_proj(t) mudou de sinal
    # se mudou, calcula a raiz
    for i in range(len(valores)-1):
        if f_proj(valores[i]) * f_proj(valores[i+1]) < 0:
            raiz = fsolve(f_proj, valores[i])
            todos.append(raiz)
    return todos
````
