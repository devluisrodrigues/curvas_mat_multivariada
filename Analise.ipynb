````python
import math as m
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
````

````python
def check(A,B):
    cond = True
    if A >= 8*m.pi:
        cond = False
    if m.sin(A/2) == 0:
        cond = False
    if m.sin(B/2) == 0:
        cond = False
    if m.sin(7*A/4) == 0:
        cond = False
    if m.sin(7*B/4) == 0:
        cond = False
    return cond
````

````python
def fazGraf():
    # VALORES DE t
    t = np.linspace(0, 4*np.pi, 1000)

    # valores para x e y para cada valor de t
    x = 5*np.cos(t) - 3*np.cos(7*t/2)
    y = 5*np.sin(t) - 3*np.sin(7*t/2)
    return x,y


#Curva
x,y = fazGraf()
plt.plot(x, y)
plt.axis('equal')
plt.show()
````

````python
def f_proj(B,K):
    value = 10 * m.sin(B/2) * m.cos((1/2)*0.8* K * m.pi) - 6 * m.sin(7*B/4) * m.cos((7/4)* 0.8* K *m.pi)
    return value
````

````python
def encontraRaiz(valores,K):
    todos = []
    # percorre todos os t em valores
    # analisa se o valor de f_proj(t) mudou de sinal
    # se mudou, calcula a raiz
    for i in range(len(valores)-1):
        #raiz = fsolve(f_proj, valores[i], args=(K))

        if f_proj(valores[i],K) * f_proj(valores[i+1],K) < 0:
            raiz = valores[i]
            todos.append(raiz)
    print(todos)
    return todos
````

````python
valores = np.arange(0, 4*m.pi, 0.001)

solucoes = encontraRaiz(valores,1)
for B in solucoes:

    u = (B + 0.8 * m.pi)/2
    v = (-B + 0.8 * m.pi)/2

    print(u)
    print(v)

    A = u + v
    
    x = 5 * m.cos(u) - 3 * m.cos(7*u/2)
    y = 5 * m.sin(u) - 3 * m.sin(7*u/2)

    pontos.append([x,y])
    plt.plot(x, y, 'ro')   

#Curva
x,y = fazGraf()
plt.plot(x, y)
plt.axis('equal')
plt.show()
````

````python
def fazGraf():
    # VALORES DE t
    t = np.linspace(0, 4*np.pi, 1000)

    # valores para x e y para cada valor de t
    x = 5*np.cos(t) - 3*np.cos(7*t/2)
    y = 5*np.sin(t) - 3*np.sin(7*t/2)

    return x,y, xl,yl



#Curva|
x,y,xl,yl = fazGraf()
plt.plot(x, y)
plt.plot(xl, yl)
plt.axis('equal')
plt.show()
````

````python
# Equacao parametrizada:
# x = 5cos(t) - 3cos(7t/2)
# y = 5sen(t) - 3sen(7t/2)

#Derivada da equacao parametrizada:
# x' = -5sen(t) + 3*7/2*sen(7t/2)
# y' = 5cos(t) - 3*7/2*cos(7t/2)

raizes_vert = []
ult = -1

for t in (np.arange(0, 4*m.pi, 0.0001)):

    xl = -5*m.sin(t) + 3*(7/2)*m.sin(7*t/2)
    yl = 5*m.cos(t) - 3*(7/2)*m.cos(7*t/2)

    # X' = 0 e Y' != 0
    if m.sqrt(xl**2) < 0.01 and m.sqrt(yl**2) > 0.01:
        if t - ult > 0.01:
            ult = t
            raizes_vert.append(t)

raizes_vert
````

````python
# Equacao parametrizada:
# x = 5cos(t) - 3cos(7t/2)
# y = 5sen(t) - 3sen(7t/2)

#Derivada da equacao parametrizada:
# x' = -5sen(t) + 3*7/2*sen(7t/2)
# y' = 5cos(t) - 3*7/2*cos(7t/2)

raizes_hor = []
ult = -1

for t in (np.arange(0, 4*m.pi, 0.0001)):

    xl = -5*m.sin(t) + 3*(7/2)*m.sin(7*t/2)
    yl = 5*m.cos(t) - 3*(7/2)*m.cos(7*t/2)

    if m.sqrt(xl**2) > 0.01 and m.sqrt(yl**2) < 0.01:
        if t - ult > 0.01:
            ult = t
            raizes_hor.append(t)

raizes_hor
````
